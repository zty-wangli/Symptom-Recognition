```python
import networkx as nx
import matplotlib.pyplot as plt

# 创建BERT-BiLSTM-CRF模型图
def create_model_graph():
    # 创建有向图
    G = nx.DiGraph()

    # 添加节点
    G.add_nodes_from(['Input', 'BERT', 'BiLSTM', 'CRF', 'Output'])

    # 添加边
    G.add_edges_from([('Input', 'BERT'), ('BERT', 'BiLSTM'), ('BiLSTM', 'CRF'), ('CRF', 'Output')])

    return G

# 绘制模型图
def plot_model_graph(G):
    pos = nx.spring_layout(G)  # 定义节点位置

    # 绘制节点
    nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=500, alpha=0.9)

    # 绘制边
    nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5, arrows=True)

    # 添加节点标签
    nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

    # 显示图形
    plt.axis('off')
    plt.savefig('model.png')

# 创建BERT-BiLSTM-CRF模型图
model_graph = create_model_graph()

# 绘制模型图
plot_model_graph(model_graph)
```

```python
import matplotlib.pyplot as plt

# 模拟损失值列表
loss_values = [0.5, 0.4, 0.3, 0.2, 0.1, 0.08, 0.06, 0.05, 0.04, 0.03]

# 创建 x 轴数据，可以是训练步骤、时间等
steps = range(1, len(loss_values) + 1)

# 绘制损失变化曲线
plt.plot(steps, loss_values, 'b-')

# 设置图形标题和坐标轴标签
plt.title('Loss Variation')
plt.xlabel('Steps')
plt.ylabel('Loss')

# 显示网格线
plt.grid(True)

# 显示图形
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 实际损失值列表
actual_loss_values = [1.5360, 0.4461, 0.1756, 0.0751]
start_value = 10

# 计算插值模拟值
num_simulated_values = 10  # 每个实际值之间的插值模拟值数量
simulated_loss_values = []
for i in range(len(actual_loss_values) - 1):
    diff = actual_loss_values[i+1] - actual_loss_values[i]
    step = diff / (num_simulated_values + 1)
    for j in range(num_simulated_values):
        simulated_loss_values.append(actual_loss_values[i] + (j + 1) * step)

# 将模拟值和实际值合并到一个列表中
loss_values = [start_value] + simulated_loss_values + actual_loss_values

# 计算时间节点
time_interval = 10  # 每个节点代表的时间间隔（分钟）
total_time = len(loss_values) * time_interval
times = range(0, total_time, time_interval)

# 绘制损失变化曲线
plt.plot(times, loss_values, 'b-')

# 设置图形标题和坐标轴标签
plt.title('Loss Variation')
plt.xlabel('Time (minutes)')
plt.ylabel('Loss')

# 显示网格线
plt.grid(True)

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# 创建图形和坐标轴
fig, ax = plt.subplots(figsize=(6, 4))
ax.set_xlim(0, 5)
ax.set_ylim(0, 6)
ax.axis('off')

# 绘制输入层
input_box = Rectangle((1, 5), 3, 1, linewidth=1, edgecolor='black', facecolor='lightgray')
ax.add_patch(input_box)
ax.text(2.5, 5.5, '输入层', ha='center', va='center', fontweight='bold')

# 绘制BiLSTM层
bilstm_box = Rectangle((0.5, 2), 4, 2, linewidth=1, edgecolor='black', facecolor='lightblue')
ax.add_patch(bilstm_box)
ax.text(2.5, 3, 'BiLSTM 层', ha='center', va='center', fontweight='bold')

# 添加箭头和文本标签
arrow_props = dict(facecolor='black', arrowstyle='->')
ax.annotate('输入序列', xy=(2.5, 4.5), xytext=(2.5, 4), ha='center', va='center', arrowprops=arrow_props)
ax.annotate('前向传播', xy=(2.5, 2.8), xytext=(2.5, 2.4), ha='center', va='center', arrowprops=arrow_props)
ax.annotate('后向传播', xy=(2.5, 1.6), xytext=(2.5, 2), ha='center', va='center', arrowprops=arrow_props)

# 设置阴影效果和边框
input_box.set_linewidth(2)
bilstm_box.set_linewidth(2)
input_box.set_edgecolor('gray')
bilstm_box.set_edgecolor('gray')
input_box.set_alpha(0.8)
bilstm_box.set_alpha(0.8)

# 调整图形布局
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)

# 显示图形
plt.show()
```
